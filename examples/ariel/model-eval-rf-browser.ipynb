```python
import pandas as pd
import numpy as np
from EIMTC.selection import train_test_split
from EIMTC.metrics import classification_report
from EIMTC.models import RandomForestClassifier
```

```python
filepath = './data/out.csv'
df = pd.read_csv(filepath, usecols=[
    'udps.src2dst_mean_clumps_len',
    'udps.src2dst_stddev_clumps_len',
    'udps.src2dst_skewness_clumps_len',
    'udps.src2dst_mean_clumps_size',
    'udps.src2dst_stddev_clumps_size',
    'udps.src2dst_skewness_clumps_size',
    'udps.src2dst_mean_clumps_bytes_per_packet',
    'udps.src2dst_stddev_clumps_bytes_per_packet',
    'udps.dst2src_mean_clumps_len',
    'udps.dst2src_stddev_clumps_len',
    'udps.dst2src_skewness_clumps_len',
    'udps.dst2src_mean_clumps_size',
    'udps.dst2src_stddev_clumps_size',
    'udps.dst2src_skewness_clumps_size',
    'udps.dst2src_mean_clumps_bytes_per_packet',
    'udps.dst2src_stddev_clumps_bytes_per_packet',
    'src2dst_mean_tls_clump_bytes', 
    'src2dst_stddev_tls_clump_bytes',
    'src2dst_mean_tls_clump_sizes',
    'src2dst_stddev_tls_clump_sizes',
    'dst2src_mean_tls_clump_bytes', 
    'dst2src_stddev_tls_clump_bytes',
    'dst2src_mean_tls_clump_sizes',
    'dst2src_stddev_tls_clump_sizes', 
    'os', 'browser'])
df
```

```python
df.dropna(inplace=True)
df
```

```python
model = RandomForestClassifier()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    df.drop(['browser', 'os'], axis=1).values, 
    df['browser'].values, 
    test_size=0.25,
    stratify=df['browser'].values,
    random_state=42
)
```

```python
model.fit(
    np.stack(x_train), 
    np.stack(y_train)
)
```

```python
predictions = model.predict(np.stack(x_test))
y_test_true = y_test

report = classification_report(y_test_true, predictions)
report
```

```python
with open('rf_clump_tls_boa2016_browser_report.txt', "w+") as f:
    f.write(report)
```
