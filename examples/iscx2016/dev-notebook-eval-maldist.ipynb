```python
import pandas as pd
import numpy as np
from EIMTC.preprocessing import OneHotEncoderEIMTC, MalDistPreprocessing
from EIMTC.selection import train_test_split
from EIMTC.models import MalDist
```

```python
filepath = './data/iscx2016_merged.csv'
cols_to_load = ['udps.n_bytes', 'udps.protocol_header_fields', 'udps.stnn_image','encapsulation', 'traffic_type']
df = pd.read_csv(filepath, usecols=cols_to_load)
df
```

```python
def preprocessing(df):
    df.dropna(inplace=True)
    MalDistPreprocessing.preprocess_features(dataframe=df)
    enc = OneHotEncoderEIMTC()
    df['enc_ohc'] = list(enc.fit_transform(df['encapsulation']).toarray())
    df['traffic_type_ohc'] = list(enc.fit_transform(df['traffic_type']).toarray())
    return df
```

```python
df = preprocessing(df)
df
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    df[['udps.n_bytes', 'udps.protocol_header_fields', 'udps.stnn_image']].values,
    df[['enc_ohc', 'traffic_type_ohc']].values,
    test_size = 0.25,
    stratify=df['traffic_type'],
    random_state=42
)
```

```python
model = MalDist(n_classes=[
    len(df['encapsulation'].unique()),
    len(df['traffic_type'].unique())
])
```

```python
epochs = 10
batch_size = 128
model.compile(
    optimizer=['adam'] *4,
    loss=['categorical_crossentropy'] *4,
    metrics=[['accuracy']] *4,
)
model.fit(
    x = [
        np.stack(x_train[:,0]),
        np.stack(x_train[:,1]), 
        np.stack(x_train[:,2])
    ], 
    y = [
        np.stack(y_train[:,0]), 
        np.stack(y_train[:,1])
    ],
    epochs=[epochs] *4,
    batch_size=[batch_size] *4,
    use_multiprocessing=[True] *4,
    workers=[4] *4,
    verbose=[2] *4,
)
```

```python
predictions_ohc = model.model.predict(
    [np.stack(x_test[:,0]),np.stack(x_test[:,1]),np.stack(x_test[:,2])], 
    verbose=1
)
predictions_os = np.argmax(predictions_ohc[0], axis=1)
predictions_browser = np.argmax(predictions_ohc[1], axis=1)
labels_os = np.argmax(np.stack(y_test[:,0]), axis=1)
labels_browser = np.argmax(np.stack(y_test[:,1]), axis=1)
```

```python
from sklearn.metrics import classification_report
report_os = classification_report(labels_os, predictions_os, target_names=df['encapsulation'].unique())
report_os
```

```python
from sklearn.metrics import classification_report
report_browser = classification_report(labels_browser, predictions_browser, target_names=df['traffic_type'].unique())
report_browser
```

```python
with open('maldist_iscx2016_encaps_report.txt', 'w+') as f:
    f.write(report_os)

with open('maldist_iscx2016_traffictype_report.txt', 'w+') as f:
    f.write(report_browser)
```
